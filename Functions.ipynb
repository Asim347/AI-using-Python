```python
def greet(name):
    print("Hello", name)

greet("Asim")
```

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result)
```

```python
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Ali")
```

```python
def student(name, age):
    print(name, age)

student(age=20, name="Sara")
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print(factorial(5))   # 120
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))   # True
```

```python
def reverse_string(s):
    return s[::-1]

print(reverse_string("Python"))
```

```python
def find_max(a, b, c):
    return max(a, b, c)

print(find_max(10, 25, 15))
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(5)
```

```python
# Lambda Function
```

```python
add = lambda a , b : a + b
print(add(3,4))
```

```python
exponent = lambda n , a : n *a 
print(exponent(3,4))
```

```python
nums = [1, 2, 3, 4]

result = list(map(lambda x: x**2, nums))
print(result)
```

```python
nums = [1, 2, 3, 4, 5, 6]

evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
data = [(1, 3), (2, 1), (4, 2)]

sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```
