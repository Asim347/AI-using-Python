```python
import numpy as np

# 1. Initialize Hyperparameters and Data
learning_rate = 0.01  # step size
epochs = 5            # number of iterations
X = np.array([1, 2, 3, 4, 5])  # input data
Y = np.array([3, 4, 2, 5, 6])  # target data
```

```python
# 2. Initialize Parameters
m, b = 0.0, 0.0  # initial parameters (slope and intercept)
n = len(X)       # number of data points

# 3. Gradient Descent Loop
for i in range(epochs):
    # Prediction: y = mx + b
    Y_pred = m * X + b
    
    # Calculate Mean Squared Error (Cost)
    cost = ((Y_pred - Y) ** 2).mean()
    
    # Calculate Derivatives (Gradients)
    # darivative_m is the partial derivative with respect to m
    darivative_m = (-2 / n) * sum(X * (Y - Y_pred))
    # darivative_b is the partial derivative with respect to b
    darivative_b = (-2 / n) * sum(Y - Y_pred)
    
    # Update Parameters
    m -= learning_rate * darivative_m
    b -= learning_rate * darivative_b
    
    # Print progress for each epoch
    print(f"Epoch {i+1}/{epochs}, Cost: {cost:.4f}, m: {m:.4f}, b: {b:.4f}")
```
