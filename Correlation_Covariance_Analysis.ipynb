```python
import pandas as pd

df = pd.read_csv("ecommerce_sales_analytics_5000.csv", parse_dates=["order_date"])
```

```python

# 2. Select only numeric columns
# Covariance and Correlation only work on numbers. 
# We create a new DataFrame 'numeric_df' containing only the relevant columns.
numeric_df = df[["quantity", "unit_price", "discount", "delivery_days", "revenue"]]

# 3. Calculate the Covariance Matrix
# Covariance shows if two variables move in the same direction. 
# Positive = move together; Negative = move in opposite directions.
# Note: Covariance values can be any number, making them hard to compare.
print("\nCovariance:\n", numeric_df.cov())
```

```python
# 4. Calculate the Correlation Matrix
# Correlation is like a "standardized" covariance. 
# It always falls between -1 and +1, making it much easier to interpret.
print("\nCorrelation:\n", numeric_df.corr())

# 5. Focus on Revenue Relationships
# This line pulls the correlation values for 'revenue' specifically,
# and sorts them from highest (most related) to lowest.
numeric_df.corr()["revenue"].sort_values(ascending=False)
```

```python
# 6. Verification
# This checks the minimum and maximum values in the correlation matrix.
# Because correlation is standardized, these values MUST be between -1 and +1.
numeric_df.corr().values.min(), numeric_df.corr().values.max()
```
