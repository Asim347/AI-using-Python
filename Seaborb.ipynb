```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("healthcare_patient_analytics_seaborn.csv")
```

```python
df = df.sample(500)
sns.scatterplot(
    data = df,
    x = "treatment_cost",
    y = "recovery_score",
    hue = "visit_type"
)
plt.title("Treament Analysis")
plt.show()
```

```python
sns.relplot(
     data = df,
    x = "treatment_cost",
    y = "recovery_score",
    hue = "visit_type",
    col = "gender"
)
```

```python
sns.set_theme(style = "darkgrid")
```

```python
plt.scatter(df["treatment_cost"], df["recovery_score"])
plt.xlabel("treatment_cost")
plt.ylabel("recovery_score")
```

```python
sns.scatterplot(
    data = df,
    x = "treatment_cost",
    y = "recovery_score"
)
```

```python
sns.histplot(
    data = df,
    x = "treatment_cost",
   bins = 30
)
plt.title("Distributio of treatment cost")
plt.show()
```

```python
sns.histplot(
    data = df,
    x = "readmission_risk",
    hue = "visit_type",
    multiple = "stack",
   bins = 30
)
plt.title("Readmission Risk")
plt.show()
```

```python
sns.kdeplot(
    data = df,
    x = "recovery_score",
    hue = "visit_type",
    fill = "True"
)
plt.title("KE of recovery Score")
plt.show()
```

```python
sns.ecdfplot(
    data = df,
    x = "length_of_stay_days",
    hue = "visit_type",
)
plt.title("ECDF of days")
plt.show()
```

```python
sns.countplot(
    data = df,
    x = "treatment_type"
)
plt.title("Distribution")
plt.show()
```

```python
sns.barplot(
    data = df,
    x = "department",
    y = "treatment_cost"
)
plt.xticks(rotation = 30)
plt.title("Distribution")
plt.show()
```

```python
sns.boxplot(
    data = df,
    x = "department",
    y = "recovery_score"
)
plt.xticks(rotation = 30)
plt.show()
```

```python
sns.violinplot(
    data = df,
    x = "department",
    y = "recovery_score"
)
plt.xticks(rotation = 30)
plt.show()
```

```python
sns.stripplot(
    data = df,
    x = "department",
    y = "recovery_score",
    jitter = True
)
plt.xticks(rotation = 30)
plt.show()
```

```python
sns.boxplot(
    data = df,
    x = "department",
    y = "recovery_score"
)
sns.stripplot(
    data = df,
    x = "department",
    y = "recovery_score",
    color = "black",
    alpha = 0.4
)
plt.xticks(rotation = 30)
plt.show()
```

```python
numeric_df = df[["length_of_stay_days","treatment_cost","recovery_score","readmission_risk"]]
```

```python
corr_matrix = numeric_df.corr()
corr_matrix
```

```python
sns.heatmap(
    corr_matrix,
    annot = True,
    cmap = "coolwarm"
)
plt.show()
```

```python
import numpy as np
mask = np.triu(np.ones_like(corr_matrix, dtype = bool))
sns.heatmap(
    corr_matrix,
    mask = mask,
    annot = True,
    cmap = "coolwarm"
)
plt.show()
```
