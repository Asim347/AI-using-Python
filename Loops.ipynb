```python
i = 4
rows = i
cols = i
matrix = []
for r in range(rows):
    rows = []
    for c in range (cols):
        rows.append(0)
    matrix.append(rows)

for row in matrix:
    print(' '.join(map(str, row)))
```

```python
rows = 5   # correct

for level in range(1, rows + 1):
    print(' ' * (rows - level) + '*' * (2 * level - 1))
```

```python
# Example 3
student = {"name": "Ali", "age": 20, "grade": "A"}

for key in student:
    print(key, ":", student[key])
```

```python
# Example 4
student_list = [
    {"name": "Ali", "age": 20, "grade": "A"},
    {"name": "Sara", "age": 22, "grade": "B"}
]

for student in student_list:
    print(f"Name: {student.get('name', '')} | Grade: {student.get('grade', '')}")
```
