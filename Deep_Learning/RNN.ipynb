```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
import pandas as pd
import numpy as np
```

```python
# Load the data
df_loaded = pd.read_csv('module_157_data.csv')

# Separate features and target
X = df_loaded.drop('target', axis=1).values
y = df_loaded['target'].values

# RESHAPE: From (2000, 10) to (2000, 10, 1)
# This adds the "feature" dimension required by RNN
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Build and compile
model = Sequential([
    # SimpleRNN is the basic recurrent layer
    SimpleRNN(32, input_shape=(10, 1)),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])  # (Adaptive Moment Estimation)
```

```python
# Train
model.fit(X, y, epochs=20, batch_size=32, validation_split=0.2)
```

```python
# Evaluate and print only the accuracy
print(f"\nFinal Accuracy: {model.evaluate(X, y, verbose=0)[1]:.4f}")
```
