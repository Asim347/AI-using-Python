---
This notebook implements the complete, dataset-agnostic preprocessing pipeline:
1. Median imputation (numerical features)  
2. Mode imputation (categorical features)  
3. Duplicate removal  
4. Label Encoding  
5. Min-Max Scaling  
6. RFE Feature Selection (~75% retention)  
7. Stratified 80:20 Split (random_state=42)


```
# ── Install dependencies ─────────────────────────────────────────
# Run this cell once if packages are missing
# !pip install pandas numpy scikit-learn xgboost joblib
```

```
# ── Imports ──────────────────────────────────────────────────────
import os, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import joblib

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedShuffleSplit

print("✓ All imports successful")
print(f"  pandas  {pd.__version__}")
print(f"  numpy   {np.__version__}")
```

## 1. Load Your Dataset
Change `DATASET_PATH` and `TARGET_COL` to match your CSV file.  
The rest of the notebook adapts automatically.


```
# ── CONFIGURE THESE TWO LINES ────────────────────────────────────
DATASET_PATH = "data/your_dataset.csv"   # ← path to your CSV file
TARGET_COL   = "role"                    # ← name of the label/target column
# ─────────────────────────────────────────────────────────────────

# Load
df = pd.read_csv(DATASET_PATH)
print(f"✓ Loaded: {DATASET_PATH}")
print(f"  Shape   : {df.shape}")
print(f"  Columns : {list(df.columns)}")
print(f"  Missing : {df.isnull().sum().sum()}")
print()
print("Class distribution:")
print(df[TARGET_COL].value_counts())
```

## 2. Detect Column Types

```
# Auto-detect numerical vs categorical columns (excluding target)
categorical_cols = [c for c in df.select_dtypes(include=["object","category"]).columns
                    if c != TARGET_COL]
numerical_cols   = [c for c in df.select_dtypes(include="number").columns
                    if c != TARGET_COL]

print(f"Numerical  ({len(numerical_cols)})  : {numerical_cols}")
print(f"Categorical({len(categorical_cols)}): {categorical_cols}")
```

## 3. Step 1+2 — Imputation | Step 3 — Duplicate Removal

```
df_clean = df.copy()
before   = len(df_clean)

# Step 1 — Median imputation for all numerical features (paper: Section IV.D bullet 1)
for col in numerical_cols:
    if df_clean[col].isnull().any():
        df_clean[col].fillna(df_clean[col].median(), inplace=True)

# Step 2 — Mode imputation for categorical features (paper: Section IV.D bullet 2)
for col in categorical_cols:
    if df_clean[col].isnull().any():
        m = df_clean[col].mode()
        df_clean[col].fillna(m[0] if not m.empty else "Unknown", inplace=True)

# Step 3 — Duplicate removal (paper: Section IV.D bullet 3)
df_clean.drop_duplicates(inplace=True)
n_removed = before - len(df_clean)

print(f"✓ Imputation complete — missing remaining: {df_clean.isnull().sum().sum()}")
print(f"✓ Duplicates removed : {n_removed}")
print(f"  Dataset shape after cleaning: {df_clean.shape}")
```

## 4. Step 4 — Label Encoding

```
cat_encoders = {}
le_target    = LabelEncoder()

# Encode all categorical feature columns
for col in categorical_cols:
    if col in df_clean.columns:
        le = LabelEncoder()
        df_clean[col] = le.fit_transform(df_clean[col].astype(str))
        cat_encoders[col] = le

# Encode the target column
df_clean[TARGET_COL] = le_target.fit_transform(df_clean[TARGET_COL].astype(str))

print(f"✓ Label Encoding applied to {len(categorical_cols)} feature column(s)")
print(f"  Target classes: {list(le_target.classes_)}")
```

## 5. Step 5 — Min-Max Scaling

```
scaler = MinMaxScaler()

num_cols_present = [c for c in numerical_cols if c in df_clean.columns]
df_clean[num_cols_present] = scaler.fit_transform(df_clean[num_cols_present])

print(f"✓ Min-Max Scaling applied to {len(num_cols_present)} numerical column(s) → [0, 1]")
```

## 6. Step 6 — RFE Feature Selection (~75%)

```
X_all = df_clean.drop(columns=[TARGET_COL])
y_all = df_clean[TARGET_COL]

total      = X_all.shape[1]
n_keep     = max(5, int(total * 0.75))   # ~75% of features, minimum 5

rfe = RFE(
    estimator=RandomForestClassifier(n_estimators=50, random_state=42),
    n_features_to_select=n_keep
)
rfe.fit(X_all, y_all)

selected_features = X_all.columns[rfe.support_].tolist()
print(f"✓ RFE complete — kept {n_keep}/{total} features (~75%)")
print(f"  Selected features: {selected_features}")

df_selected = df_clean[selected_features + [TARGET_COL]]
```

## 7. Step 7 — Stratified 80:20 Split

```
X = df_selected.drop(columns=[TARGET_COL]).values
y = df_selected[TARGET_COL].values

sss = StratifiedShuffleSplit(n_splits=1, test_size=0.20, random_state=42)
train_idx, test_idx = next(sss.split(X, y))

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

print(f"✓ Split complete — random_state=42 | stratified")
print(f"  Train : {len(X_train)} samples")
print(f"  Test  : {len(X_test)}  samples")
print(f"  Ratio : 80:20")
print()

# Verify class balance
import collections
print("Train class counts:", dict(collections.Counter(y_train)))
print("Test  class counts:", dict(collections.Counter(y_test)))
```

## 8. Save Preprocessor for Later Notebooks

```
os.makedirs("models", exist_ok=True)

# Bundle everything into a dict for easy reuse
preprocessor = {
    "scaler":           scaler,
    "le_target":        le_target,
    "cat_encoders":     cat_encoders,
    "selected_features":selected_features,
    "numerical_cols":   numerical_cols,
    "categorical_cols": categorical_cols,
    "target_col":       TARGET_COL,
    "class_names":      list(le_target.classes_),
    "n_features_total": total,
    "n_features_selected": n_keep,
}

joblib.dump(preprocessor, "models/preprocessor.joblib")
print("✓ Preprocessor saved → models/preprocessor.joblib")
print()
print("=== PREPROCESSING SUMMARY ===")
print(f"  Original shape      : {df.shape}")
print(f"  After cleaning      : {df_clean.shape}")
print(f"  Duplicates removed  : {n_removed}")
print(f"  Features selected   : {n_keep}/{total}")
print(f"  Train / Test        : {len(X_train)} / {len(X_test)}")
print(f"  Classes             : {list(le_target.classes_)}")
```
