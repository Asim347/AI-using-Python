```python
import numpy as np
import pandas as pd

df = pd.read_csv("telecom_customer_churn_feature_engineering.csv")
```

```python
len(df)
```

```python
df_shuffled = df.sample(frac = 1, random_state = 42)
train_size = int(0.8 * len(df_shuffled))

train_df = df_shuffled.iloc[:train_size]
test_df = df_shuffled.iloc[train_size:]

len(train_df), len(test_df)
```

```python
set(train_df.index).intersection(set(test_df.index))
```

```python
# Scaling before split (Data Leakage Example)
inc = df["monthly_income"]
df["income_scaled_wrong"] = (inc - inc.mean()) / inc.std()

df[["monthly_income", "income_scaled_wrong"]].head()
```

```python
# Separate Features & Target
X_train = train_df.drop("churn", axis=1)
y_train = train_df["churn"]

X_test = test_df.drop("churn", axis=1)
y_test = test_df["churn"]

X_train.shape, X_test.shape

# Target is safely isolated
# No leakage risk
```
