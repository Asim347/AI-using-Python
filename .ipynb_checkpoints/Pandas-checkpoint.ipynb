```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 35],
    'City': ['New York', 'Paris', 'London', 'Berlin'],
    'Salary': [70000, 80000, 60000, 90000]
}
df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

# 1. Label-based selection (.loc)
# Logic: df.loc[row_labels, column_names]
subset_loc = df.loc[1:3, ['Name', 'Salary']]
print("\nSubset with .loc:\n", subset_loc)

# 2. Integer-based selection (.iloc)
# Logic: df.iloc[row_index, column_index]
subset_iloc = df.iloc[0:2, 1:3]
print("\nSubset with .iloc:\n", subset_iloc)
```

```python
high_salary = df[df['Salary'] > 75000]
print("High salary is:", high_salary)
```

```python
 # Safe assignment using .loc
df.loc[df['Age'] < 30, 'Salary'] = 65000
print("Low Age Salary Change:\n", df.loc[df['Age'] < 30, 'Salary'])
```

```python
df = pd.read_csv(
    "Amazon.csv",
    parse_dates=["OrderDate"],
    dtype={
        "Category": "category",
        "Region": "category"
    }
)
```

```python
df.info()
```

```python
clean_df = df[df["TotalAmount"] > 0]   # remove invalid rows

subset = clean_df[["OrderDate", "Category", "TotalAmount", "Country"]]
```

```python
subset.to_excel("subset_sales.xlsx", sheet_name='sales')

subset.to_json("subset_sales.json",
               orient="records",
               date_format="iso")

print("Files exported!")
subset.info()
```

```python
# 1. Create the DataFrame with messy strings
df = pd.DataFrame({
    'Category': [' Electronics ', 'furniture', 'Home-Appliances', ' electronics', 'FURNITURE ']
})

print(df)
```

```python
# 2. Strip spaces and convert to lowercase
df['Category_clean'] = df['Category'].str.strip().str.lower()

print(df)
```

```python
# 3. Replace hyphens with space
df['Category_clean'] = df['Category_clean'].str.replace('-', ' ')

print(df)
```

```python
customers = pd.DataFrame({
    'Customer_id' : [1,2,3],
    'Name' : ['ali', 'Asim', 'khan']
})

orders = pd.DataFrame({
    'ID' : [101,102,103,104],
    'Customer_id' : [1,2,2,3],
    'Amount' : [200,300,400,500]
})
```

```python
df_inner = pd.merge(customers, orders, on = 'Customer_id', how = 'inner')
print("Merge dataframe\n: ", df_inner)
```

```python
df_inner = pd.merge(customers, orders, on = 'Customer_id', how = 'outer')
print("Merge dataframe\n: ", df_inner)
```
