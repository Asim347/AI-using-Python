```python
# 1. Open the file in 'w+' mode (Write and Read)
# This creates 'demo.txt' or overwrites it if it already exists
file = open("demo.txt", "w+")

# 2. Write content to the file
file.write("this is test of new line")

# 3. Move the cursor back to the beginning of the file
# After writing, the cursor is at the end, so read() would return nothing
file.seek(0)

# 4. Read and print the content
print(file.read())

# 5. Always close the file to save changes and free memory
file.close()
```

```python
file1 = open("demo.txt", "r")
print(file1.read())
file1.close()
```

```python
file1 = open("demo.txt", "w")
file1.write("This is another test.")
file1.close()
```

```python
file1 = open("demo.txt", "a")
file1.write("\nNew line added!")
file1.close()
```

```python
with open("demo.txt","r") as file:
    data = file.read()
    print(data)
```

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    # Triggered if user enters 0
    print("You can't divide by zero!")
except ValueError:
    # Triggered if user enters text instead of a number
    print("That's not a number!")
except Exception as e:
    # Catch-all for any other unexpected errors
    print(f"An unexpected error occurred: {e}")
```

```python
try:
    # 1. The code we want to attempt
    num = int(input("Enter a number: "))
    result = 10 / num
    print(f"Result: {result}")

except ZeroDivisionError:
    # 2. Handles division by zero
    print("Error: You can't divide by zero!")

except ValueError:
    # 3. Handles non-numeric input (like letters)
    print("Error: That's not a valid number!")

except Exception as e:
    # 4. A safety net for any other unexpected errors
    print(f"An unexpected error occurred: {e}")

else:
    # 5. Runs ONLY if no errors occurred in the 'try' block
    print("Calculation successful! No errors found.")

finally:
    # 6. Runs EVERY TIME, no matter what happened above
    # Perfect for closing files or cleaning up
    print("Task execution complete. (Finally block)")
```
