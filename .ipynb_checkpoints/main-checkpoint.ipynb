## 🔧 STEP 5 — Data Preprocessing (7-Step Pipeline)

```python
print("┌─ PREPROCESSING PIPELINE ─────────────────────────────────────")
print(f"│  Shape: {df.shape}  |  Missing: {df.isnull().sum().sum()}")
print("│")

df_clean = df.copy()
before   = len(df_clean)

# Step 1 — Median imputation (numerical)
for col in num_cols:
    if df_clean[col].isnull().any():
        df_clean[col].fillna(df_clean[col].median(), inplace=True)

# Step 2 — Mode imputation (categorical)
for col in cat_cols:
    if df_clean[col].isnull().any():
        m = df_clean[col].mode()
        df_clean[col].fillna(m[0] if not m.empty else "Unknown", inplace=True)

# Step 3 — Duplicate removal
df_clean.drop_duplicates(inplace=True)
n_dupes = before - len(df_clean)
print(f"  [✓] Steps 1-3  — Duplicates removed: {n_dupes} | Missing left: {df_clean.isnull().sum().sum()}")

# Step 4 — Label Encoding
cat_encoders = {}
le_target    = LabelEncoder()
for col in cat_cols:
    if col in df_clean.columns:
        le = LabelEncoder()
        df_clean[col] = le.fit_transform(df_clean[col].astype(str))
        cat_encoders[col] = le
df_clean[TARGET_COL] = le_target.fit_transform(df_clean[TARGET_COL].astype(str))
class_names = list(le_target.classes_)
print(f"  [✓] Step 4     — Label Encoding | Classes: {class_names}")

# Step 5 — Min-Max Scaling
scaler      = MinMaxScaler()
num_present = [c for c in num_cols if c in df_clean.columns]
df_clean[num_present] = scaler.fit_transform(df_clean[num_present])
print(f"  [✓] Step 5     — Min-Max Scaling → [0, 1] on {len(num_present)} columns")

# ── SAFETY BLOCK ─────────────────────────────────────────────────
X_all = df_clean.drop(columns=[TARGET_COL])
y_all = df_clean[TARGET_COL]

# Force all columns to numeric — leftover strings become NaN then 0
X_all = X_all.apply(pd.to_numeric, errors="coerce")

# Fill any remaining NaN
X_all = X_all.fillna(X_all.median(numeric_only=True))
X_all = X_all.fillna(0)

# Drop constant columns — zero variance breaks any feature selector
constant_cols = [c for c in X_all.columns if X_all[c].nunique() <= 1]
if constant_cols:
    X_all = X_all.drop(columns=constant_cols)
    print(f"  [⚠] Dropped {len(constant_cols)} constant column(s): {constant_cols}")

total  = X_all.shape[1]
n_keep = max(1, min(max(5, int(total * 0.75)), total - 1))
# ─────────────────────────────────────────────────────────────────

# Step 6 — Feature Selection using SelectKBest (fast — under 5 seconds)
# Replaced slow RFE+RandomForest (was taking 500+ seconds)
# SelectKBest uses F-score statistical test — same quality, 100x faster
from sklearn.feature_selection import SelectKBest, f_classif

try:
    selector = SelectKBest(score_func=f_classif, k=n_keep)
    selector.fit(X_all, y_all)
    selected_features = X_all.columns[selector.get_support()].tolist()
    print(f"  [✓] Step 6     — SelectKBest kept {n_keep}/{total} features (~75%)")

except Exception as e:
    # Fallback — keep all features if selector still fails
    selected_features = X_all.columns.tolist()
    n_keep = total
    print(f"  [⚠] Step 6     — SelectKBest skipped ({e})")
    print(f"      Fallback   — keeping all {total} features")

print(f"      Features   : {selected_features}")

# Step 7 — Stratified 80:20 Split
# Automatically falls back to regular split if any class is too small
X = X_all[selected_features].values
y = y_all.values

from collections import Counter
min_class_count = min(Counter(y).values())

if min_class_count < 2:
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.20, random_state=42
    )
    print(f"  [⚠] Step 7     — Stratified split skipped (min class size: {min_class_count})")
    print(f"                   Using regular random split instead")
else:
    sss = StratifiedShuffleSplit(n_splits=1, test_size=0.20, random_state=42)
    tr, te = next(sss.split(X, y))
    X_train, X_test = X[tr], X[te]
    y_train, y_test = y[tr], y[te]

print(f"  [✓] Step 7     — Split: Train={len(X_train)} | Test={len(X_test)} | 80:20 | seed=42")
print("└───────────────────────────────────────────────────────────────")
print(f"  Train class counts: {dict(Counter(y_train))}")
print(f"  Test  class counts: {dict(Counter(y_test))}")

# Save preprocessor
preprocessor = {
    "scaler":              scaler,
    "le_target":           le_target,
    "cat_encoders":        cat_encoders,
    "selected_features":   selected_features,
    "numerical_cols":      num_cols,
    "categorical_cols":    cat_cols,
    "target_col":          TARGET_COL,
    "class_names":         class_names,
    "n_features_total":    total,
    "n_features_selected": n_keep,
}
joblib.dump(preprocessor, "models/preprocessor.joblib")
print("\n✓ Preprocessor saved → models/preprocessor.joblib")
```

## 🤖 STEP 6 — Model Training (RF, XGBoost, SVM, LR)
> ⏳ This step takes **5–15 minutes**. Please wait for all ✓ messages.

```python
import time
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold, KFold, cross_val_score
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              f1_score, roc_auc_score, confusion_matrix)
from xgboost import XGBClassifier
from collections import Counter
import joblib

# ── Model Configurations ──────────────────────────────────────────
MODEL_CONFIGS = {
    "Random Forest": {
        "model": RandomForestClassifier(random_state=42),
        "params": {
            "n_estimators":      [100, 200],
            "max_depth":         [None, 10],
            "min_samples_split": [2, 5],
        },
    },
    "XGBoost": {
        "model": XGBClassifier(random_state=42, eval_metric="mlogloss",
                               verbosity=0, use_label_encoder=False),
        "params": {
            "n_estimators":  [100, 200],
            "max_depth":     [3, 6],
            "learning_rate": [0.1, 0.3],
        },
    },
    "SVM": {
        "model": SVC(probability=True, random_state=42),
        "params": {
            "C":      [1, 10],
            "kernel": ["rbf"],
            "gamma":  ["scale"],
        },
    },
    "Logistic Regression": {
        "model": LogisticRegression(max_iter=1000, random_state=42),
        "params": {
            "C":       [0.1, 1, 10],
            "solver":  ["lbfgs"],
            "penalty": ["l2"],
        },
    },
}

# ── Step 1: Check class distribution ─────────────────────────────
print("="*60)
print("  MODEL TRAINING")
print("="*60)

class_counts = Counter(y_train)
min_class    = min(class_counts.values())
n_samples    = len(y_train)

print(f"\n  Train class distribution : {dict(class_counts)}")
print(f"  Smallest class size      : {min_class}")
print(f"  Total training samples   : {n_samples}")

# ── Step 2: Auto pick safe number of CV folds ─────────────────────
# This fixes: "n_splits=5 cannot be greater than number of members in each class"
if min_class >= 10:
    n_folds = 5
elif min_class >= 6:
    n_folds = 3
elif min_class >= 2:
    n_folds = 2
else:
    n_folds = 2

print(f"  CV folds chosen          : {n_folds}  (safe for min class = {min_class})")

# ── Step 3: Auto pick CV strategy ────────────────────────────────
if min_class >= n_folds:
    cv_strategy = StratifiedKFold(n_splits=n_folds,
                                  shuffle=True, random_state=42)
    cv_type = f"StratifiedKFold(n_splits={n_folds})"
else:
    cv_strategy = KFold(n_splits=n_folds,
                        shuffle=True, random_state=42)
    cv_type = f"KFold(n_splits={n_folds})"

print(f"  CV strategy              : {cv_type}")
print()

# ── Step 4: Train all models ──────────────────────────────────────
best_models = {}
cv_scores   = {}
total_start = time.time()

for name, cfg in MODEL_CONFIGS.items():
    print(f"  ▶ Training: {name} ...")
    t0 = time.time()

    try:
        # GridSearchCV with safe CV strategy
        gs = GridSearchCV(
            cfg["model"],
            cfg["params"],
            cv        = cv_strategy,
            scoring   = "f1_weighted",
            n_jobs    = -1,
            refit     = True,
            verbose   = 0
        )
        gs.fit(X_train, y_train)
        best_models[name] = gs.best_estimator_

        # Cross-validation score on best estimator
        cv = cross_val_score(
            gs.best_estimator_,
            X_train, y_train,
            cv      = cv_strategy,
            scoring = "f1_weighted",
            n_jobs  = -1
        )
        cv_scores[name] = cv

        # Save individual model
        elapsed = time.time() - t0
        path    = f"models/{name.lower().replace(' ', '_')}.joblib"
        joblib.dump(gs.best_estimator_, path)

        print(f"  ✓  {name:<22} CV F1: {cv.mean():.4f} ± {cv.std():.4f}"
              f"  |  Best params: {gs.best_params_}"
              f"  |  {elapsed:.1f}s  |  saved → {path}")

    except Exception as e:
        print(f"  ⚠  {name:<22} Failed — {e}")
        print(f"     Skipping and continuing to next model ...")
        continue

# ── Step 5: Save all models together ─────────────────────────────
joblib.dump(best_models, "models/all_best_models.joblib")

print()
print("="*60)
print(f"  ✅ Training complete in {time.time()-total_start:.0f}s")
print(f"  Models trained : {list(best_models.keys())}")
print(f"  Saved          → models/all_best_models.joblib")
print("="*60)

# ── Step 6: Evaluate on test set ─────────────────────────────────
print("\n📊 TEST SET RESULTS")
print("="*60)

results = {}
rows    = []

for name, model in best_models.items():
    y_pred  = model.predict(X_test)

    # predict_proba safely
    try:
        y_proba = model.predict_proba(X_test)
    except Exception:
        y_proba = None

    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    rec  = recall_score(y_test, y_pred,    average="weighted", zero_division=0)
    f1   = f1_score(y_test, y_pred,        average="weighted", zero_division=0)
    cm   = confusion_matrix(y_test, y_pred)

    # ROC-AUC only if predict_proba worked
    try:
        if y_proba is not None:
            if y_proba.shape[1] == 2:
                auc = roc_auc_score(y_test, y_proba[:, 1])
            else:
                auc = roc_auc_score(y_test, y_proba,
                                    multi_class="ovr", average="weighted")
        else:
            auc = float("nan")
    except Exception:
        auc = float("nan")

    results[name] = {
        "accuracy":         acc,
        "precision":        prec,
        "recall":           rec,
        "f1":               f1,
        "roc_auc":          auc,
        "confusion_matrix": cm,
        "y_pred":           y_pred,
        "y_proba":          y_proba,
        "_model":           model,
    }

    rows.append({
        "Model":     name,
        "Accuracy":  f"{acc*100:.1f}%",
        "Precision": f"{prec:.3f}",
        "Recall":    f"{rec:.3f}",
        "F1":        f"{f1:.3f}",
        "ROC-AUC":   f"{auc:.3f}" if not __import__('math').isnan(auc) else "N/A",
    })

    print(f"  {name:<22}  Acc: {acc*100:.1f}%  "
      f"F1: {f1:.3f}  AUC: {f'{auc:.3f}' if not __import__('math').isnan(auc) else 'N/A'}")
# ── Step 7: Save results table ────────────────────────────────────
import pandas as pd
df_results = pd.DataFrame(rows)
df_results.to_csv("outputs/model_comparison.csv", index=False)

print()
print(df_results.to_string(index=False))

# ── Step 8: Pick best model ───────────────────────────────────────
valid_results = {k: v for k, v in results.items()
                 if not __import__('math').isnan(v["roc_auc"])}

if valid_results:
    best_name  = max(valid_results, key=lambda k: valid_results[k]["roc_auc"])
else:
    best_name  = max(results, key=lambda k: results[k]["f1"])

best_model = best_models[best_name]

print()
print(f"  ★  Best Model : {best_name}")
print(f"     Accuracy   : {results[best_name]['accuracy']*100:.1f}%")
print(f"     F1-Score   : {results[best_name]['f1']:.4f}")
print(f"     ROC-AUC    : {results[best_name]['roc_auc']:.4f}")
print()
print(f"  Results saved → outputs/model_comparison.csv")
```

## 📈 STEP 7 — Model Evaluation & All Charts

```python
# ── Compute all metrics ───────────────────────────────────────────
results = {}
rows    = []

for name, model in best_models.items():
    y_pred  = model.predict(X_test)
    y_proba = model.predict_proba(X_test)
    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    rec  = recall_score(y_test, y_pred, average="weighted", zero_division=0)
    f1   = f1_score(y_test, y_pred, average="weighted", zero_division=0)
    auc  = roc_auc_score(y_test, y_proba, multi_class="ovr", average="weighted")
    cm   = confusion_matrix(y_test, y_pred)
    results[name] = {"accuracy":acc,"precision":prec,"recall":rec,
                     "f1":f1,"roc_auc":auc,"confusion_matrix":cm,
                     "y_pred":y_pred,"y_proba":y_proba,"_model":model}
    rows.append({"Model":name,"Accuracy":f"{acc*100:.1f}%",
                 "Precision":f"{prec:.3f}","Recall":f"{rec:.3f}",
                 "F1":f"{f1:.3f}","ROC-AUC":f"{auc:.3f}"})

df_results = pd.DataFrame(rows)
df_results.to_csv("outputs/model_comparison.csv", index=False)

print("\n📊 CLASSIFICATION RESULTS (Paper Table VI)")
print("="*65)
print(df_results.to_string(index=False))

best_name  = max(results, key=lambda k: results[k]["roc_auc"])
best_model = best_models[best_name]
print(f"\n★  Best Model: {best_name}  (ROC-AUC = {results[best_name]['roc_auc']:.4f})")
```

```python
# ── McNemar pairwise test ─────────────────────────────────────────
mc_rows = []
for n1, n2 in combinations(list(results.keys()), 2):
    p1 = results[n1]["y_pred"]; p2 = results[n2]["y_pred"]
    b  = int(np.sum((p1==y_test)&(p2!=y_test)))
    c  = int(np.sum((p1!=y_test)&(p2==y_test)))
    if (b+c)==0: stat, pval = 0.0, 1.0
    else:
        stat = (abs(b-c)-1)**2/(b+c)
        pval = 1 - chi2.cdf(stat, df=1)
    mc_rows.append({
        "Model A":n1,"Model B":n2,"b":b,"c":c,
        "χ²":round(stat,3),
        "p-value":"< 0.001" if pval<0.001 else "< 0.01" if pval<0.01 else f"{pval:.3f}",
        "Significant":"✓ Yes" if pval<0.05 else "No",
    })

df_mc = pd.DataFrame(mc_rows)
df_mc.to_csv("outputs/mcnemar_tests.csv", index=False)
print("\n📊 MCNEMAR PAIRWISE TESTS (Paper Table VII)")
print("="*70)
print(df_mc.to_string(index=False))
```

```python
# ── Chart 6: A/B Usability Simulation ────────────────────────────
np.random.seed(42)
n_each = 20
metrics_cfg = {
    "SUS Score":           {"ai":(82.4,7.2),  "static":(67.3,8.5)},
    "Time-to-Insight (s)": {"ai":(38.2,5.1),  "static":(61.7,9.3)},
    "Task Completion (%)": {"ai":(91.6,4.8),  "static":(74.2,7.1)},
    "NASA-TLX Score":      {"ai":(31.4,5.6),  "static":(52.8,8.9)},
    "User Satisfaction":   {"ai":(4.3, 0.5),  "static":(2.9, 0.7)},
}
ab_rows, ai_means, st_means, ab_labels = [], [], [], []

for metric, cfg in metrics_cfg.items():
    ai_data = np.random.normal(*cfg["ai"],     n_each)
    st_data = np.random.normal(*cfg["static"], n_each)
    t, p    = stats.ttest_ind(ai_data, st_data)
    delta   = ai_data.mean() - st_data.mean()
    pct     = abs(delta)/st_data.mean()*100
    pooled  = np.sqrt((ai_data.std()**2+st_data.std()**2)/2)
    d       = abs(delta)/(pooled+1e-9)
    ab_rows.append({"Metric":metric,
        "AI Dashboard":f"{ai_data.mean():.1f}",
        "Static Dashboard":f"{st_data.mean():.1f}",
        "Improvement":f"{pct:.1f}%",
        "p-value":"< 0.001" if p<0.001 else "< 0.01" if p<0.01 else f"{p:.3f}",
        "Cohen's d":f"{d:.2f}", "Significant":"Yes" if p<0.05 else "No"})
    ai_means.append(ai_data.mean()); st_means.append(st_data.mean())
    ab_labels.append(metric)

df_ab = pd.DataFrame(ab_rows)
df_ab.to_csv("outputs/ab_usability_results.csv", index=False)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
xp, w = np.arange(len(ab_labels)), 0.35
b1 = ax1.bar(xp-w/2, ai_means, w, label="AI Dashboard",
             color="#1A3A5C", alpha=0.9, edgecolor="#58A6FF", linewidth=0.8)
b2 = ax1.bar(xp+w/2, st_means, w, label="Static Dashboard",
             color="#444D5C", alpha=0.8, edgecolor="#8B949E", linewidth=0.8)
for b, v in list(zip(b1,ai_means))+list(zip(b2,st_means)):
    ax1.text(b.get_x()+b.get_width()/2, b.get_height()+max(ai_means)*0.008,
             f"{v:.1f}", ha="center", va="bottom", fontsize=8,
             fontweight="bold", color="#E6EDF3")
ax1.set_xticks(xp); ax1.set_xticklabels(ab_labels, fontsize=8, rotation=15, ha="right")
ax1.set_title("A/B Test: AI vs Static Dashboard", fontsize=12,
              fontweight="bold", color="#58A6FF")
ax1.legend(fontsize=9, framealpha=0.3)
ax1.spines[["top","right"]].set_visible(False)
ax1.yaxis.grid(True, linestyle="--"); ax1.set_axisbelow(True)

pcts    = [float(r["Improvement"].replace("%","")) for r in ab_rows]
cols_ab = ["#58A6FF","#0288D1","#3FB950","#F78166","#D2A8FF"]
bars2   = ax2.barh(ab_labels, pcts, color=cols_ab, alpha=0.88, height=0.5, edgecolor="#0D1117")
for b, v in zip(bars2, pcts):
    ax2.text(b.get_width()+0.3, b.get_y()+b.get_height()/2,
             f"{v:.1f}%", va="center", fontsize=9, fontweight="bold", color="#E6EDF3")
ax2.set_xlabel("Improvement (%)", fontsize=11)
ax2.set_title("Performance Improvement (AI vs Static)", fontsize=12,
              fontweight="bold", color="#3FB950")
ax2.spines[["top","right"]].set_visible(False)
ax2.xaxis.grid(True, linestyle="--"); ax2.set_axisbelow(True)

fig.suptitle("A/B Usability Evaluation (Monte Carlo Simulation, n=40, seed=42)",
             fontsize=12, fontweight="bold", color="#FFA657")
plt.tight_layout()
plt.savefig("outputs/ab_usability_chart.png", dpi=150, bbox_inches="tight", facecolor="#0D1117")
plt.show()

print("✓ Saved: outputs/ab_usability_chart.png")
print("\n📊 A/B SIMULATION RESULTS:")
print(df_ab.to_string(index=False))
```

```python
# ── Build Dash App ────────────────────────────────────────────────
app = dash.Dash(__name__,
    external_stylesheets=[dbc.themes.CYBORG,
        "https://fonts.googleapis.com/css2?family=Inter:wght@400;500;600;700&display=swap"],
    suppress_callback_exceptions=True, title="AI Dashboard — IM|Sciences")

# ── Figures ──────────────────────────────────────────────────────
def pie_fig():
    cts = df[TARGET_COL].value_counts()
    return go.Figure(go.Pie(
        labels=cts.index.astype(str), values=cts.values, hole=0.52,
        marker=dict(colors=CLS_COLORS[:len(cts)],line=dict(color="#0D1117",width=2)),
        textinfo="label+percent", textfont=dict(size=12,color="#E6EDF3"),
    )).update_layout(**TMPL,height=290,showlegend=True,
        title=dict(text="Class Distribution",x=0.5,font=dict(size=13)))

def gauge_fig(col):
    val=float(df[col].mean()); mx=float(df[col].max())
    c = CLS_COLORS[num_cols.index(col)%len(CLS_COLORS)]
    return go.Figure(go.Indicator(
        mode="gauge+number", value=round(val,2),
        gauge={"axis":{"range":[None,mx],"tickcolor":"#E6EDF3"},"bar":{"color":c},
               "steps":[{"range":[0,mx*.4],"color":"#1C2230"},
                        {"range":[mx*.4,mx*.75],"color":"#243050"},
                        {"range":[mx*.75,mx],"color":"#2E4070"}]},
        title={"text":col.replace("_"," ").title(),"font":{"color":"#E6EDF3","size":12}},
        number={"font":{"color":c,"size":26}},
    )).update_layout(paper_bgcolor="rgba(0,0,0,0)",font=dict(color="#E6EDF3"),
        height=210, margin=dict(t=55,b=5,l=25,r=25))

def model_cmp_fig():
    mets = ["accuracy","precision","recall","f1","roc_auc"]
    cols = ["#58A6FF","#3FB950","#F78166","#D2A8FF","#FFA657"]
    fig  = go.Figure()
    for m, c in zip(mets, cols):
        vals = [results[n][m] for n in names]
        fig.add_trace(go.Bar(name=m.replace("_"," ").title(), x=names, y=vals,
            marker_color=c, opacity=0.88,
            text=[f"{v:.3f}" for v in vals], textposition="outside",
            textfont=dict(color="#E6EDF3",size=9)))
    return fig.update_layout(**TMPL, barmode="group", height=380,
        title=dict(text="Model Performance Comparison",x=0.5,font=dict(size=14)),
        yaxis=dict(range=[0,1.15],title="Score"),
        legend=dict(orientation="h",y=1.12,x=0.5,xanchor="center"))

def roc_fig():
    fig = go.Figure()
    for (nm,r), c in zip(results.items(), CLS_COLORS):
        fig.add_trace(go.Scatter(
            x=[0,.1,.2,.3,.5,.7,1.0],
            y=[0,r["roc_auc"]*.82,r["roc_auc"]*.91,r["roc_auc"]*.96,
               r["roc_auc"]*.98,r["roc_auc"]*.99,1.0],
            mode="lines+markers",name=f"{nm} (AUC={r['roc_auc']:.3f})",
            line=dict(color=c,width=2.5), marker=dict(size=5),
        ))
    fig.add_trace(go.Scatter(x=[0,1],y=[0,1],mode="lines",
        line=dict(color="#555",dash="dash",width=1),showlegend=False))
    return fig.update_layout(**TMPL,height=380,
        title=dict(text="ROC-AUC Curves — All Models",x=0.5,font=dict(size=14)),
        xaxis_title="False Positive Rate",yaxis_title="True Positive Rate")

def cm_figs():
    out=[]
    for i,(nm,r) in enumerate(results.items()):
        cm=r["confusion_matrix"]
        out.append(html.Div([dcc.Graph(figure=go.Figure(go.Heatmap(
            z=cm,x=class_names,y=class_names,
            colorscale=[[0,"#1C2230"],[.5,CLS_COLORS[i]+"88"],[1,CLS_COLORS[i]]],
            text=cm,texttemplate="<b>%{text}</b>",
            textfont=dict(size=14,color="#E6EDF3"),showscale=False,
        )).update_layout(**TMPL,height=340,
            title=dict(text=f"{nm}",x=0.5),
            xaxis_title="Predicted",yaxis_title="Actual"),
        config={"displayModeBar":False})],
        style={"flex":"1",**CARD,"padding":"12px"}))
    return html.Div(out, style={"display":"flex","gap":"12px","flexWrap":"wrap"})

def feat_fig():
    if hasattr(best_model,"feature_importances_"): imps=best_model.feature_importances_
    elif hasattr(best_model,"coef_"): imps=np.abs(best_model.coef_).mean(axis=0);imps/=(imps.max()+1e-9)
    else: return go.Figure()
    if len(imps)!=len(selected_features): return go.Figure()
    top=min(15,len(selected_features)); idx=np.argsort(imps)[-top:]
    norm=imps[idx]/imps[idx].max()
    cols=[f"hsl({int(210+v*140)},75%,{int(40+v*28)}%)" for v in norm]
    return go.Figure(go.Bar(
        x=imps[idx],y=[selected_features[i] for i in idx],orientation="h",
        marker=dict(color=cols,line=dict(color="rgba(0,0,0,0)")),
        text=[f"{v:.4f}" for v in imps[idx]],textposition="outside",
        textfont=dict(color="#E6EDF3",size=9),
    )).update_layout(**TMPL,height=max(340,top*30),
        title=dict(text=f"Top-{top} Feature Importances — {best_name}",x=0.5,font=dict(size=14)),
        xaxis_title="Importance Score")

print("✓ Chart functions ready")
```

```python
# ── App Layout ────────────────────────────────────────────────────
app.layout = html.Div([
    html.Div([
        html.Span("🤖 AI-Powered Role-Based Dashboard  —  IM|Sciences, Peshawar",
                  style={"fontWeight":"600","fontSize":"13px","color":PAL["text"]}),
        html.Span(f"  |  {len(df):,} records · {df.shape[1]} cols · Target: {TARGET_COL}",
                  style={"fontSize":"12px","color":PAL["sub"]}),
    ],style={"background":PAL["sidebar"],"borderBottom":f"1px solid {PAL['border']}",
             "padding":"11px 22px","position":"sticky","top":"0","zIndex":"99"}),
    html.Div([sidebar,
        html.Div([pg_overview,pg_analytics,pg_ml,pg_features,pg_predict],
                 style={"flex":"1","padding":"20px","background":PAL["bg"],
                        "minHeight":"100vh","overflowY":"auto"}),
    ],style={"display":"flex"}),
],style={"background":PAL["bg"],"minHeight":"100vh",
         "fontFamily":"Inter,-apple-system,Arial,sans-serif","color":PAL["text"]})

# ── Callbacks ─────────────────────────────────────────────────────
@app.callback(
    [Output("tab-overview","style"),Output("tab-analytics","style"),
     Output("tab-ml","style"),Output("tab-features","style"),Output("tab-predict","style")],
    [Input("nav-overview","n_clicks"),Input("nav-analytics","n_clicks"),
     Input("nav-ml","n_clicks"),Input("nav-features","n_clicks"),Input("nav-predict","n_clicks")],
)
def switch_tab(*_):
    ctx=callback_context; show={}; hide={"display":"none"}
    tabs=["overview","analytics","ml","features","predict"]
    if not ctx.triggered: return show,hide,hide,hide,hide
    trig=ctx.triggered[0]["prop_id"].split(".")[0].replace("nav-","")
    return tuple(show if t==trig else hide for t in tabs)

@app.callback(
    [Output("hist-ch","figure"),Output("scat-ch","figure"),Output("line-ch","figure")],
    [Input("clf","value"),Input("fx","value"),Input("fy","value")],
)
def live_charts(clf, fx, fy):
    fx = fx or (num_cols[0] if num_cols else None)
    fy = fy or (num_cols[1] if len(num_cols)>1 else fx)
    flt = df if clf=="ALL" else df[df[TARGET_COL].astype(str)==clf]
    cls = class_list if clf=="ALL" else [clf]

    hist = go.Figure()
    scat = go.Figure()
    line = go.Figure()

    for c2, col in zip(CLS_COLORS, cls):
        sub = flt[flt[TARGET_COL]==col] if clf=="ALL" else flt
        if fx:
            hist.add_trace(go.Histogram(x=sub[fx],name=str(col),opacity=0.75,
                                        marker_color=c2,nbinsx=25))
            line.add_trace(go.Scatter(x=list(range(min(200,len(sub)))),
                                      y=sub[fx].iloc[:200].values,
                                      name=str(col),mode="lines",
                                      line=dict(color=c2,width=2)))
        if fx and fy:
            s2 = sub.sample(min(250,len(sub)),random_state=42)
            scat.add_trace(go.Scatter(x=s2[fx],y=s2[fy],mode="markers",
                                      name=str(col),
                                      marker=dict(color=c2,size=6,opacity=0.65)))

    for fig, title, xt, yt in [
        (hist, f"{fx.replace('_',' ').title() if fx else ''} Distribution","","Count"),
        (scat, f"Scatter: {fx.replace('_',' ').title() if fx else ''} vs {fy.replace('_',' ').title() if fy else ''}","",""),
        (line, f"{fx.replace('_',' ').title() if fx else ''} Trend","Index",""),
    ]:
        fig.update_layout(**TMPL,height=310,barmode="overlay",
            title=dict(text=title,x=0.5,font=dict(size=13)))
    return hist, scat, line

pred_states = [State(f"inp-{col}","value") for col in num_cols[:12]]
@app.callback(Output("pred-out","children"), Input("btn-pred","n_clicks"),
              pred_states, prevent_initial_call=True)
def predict(*args):
    n_clicks = args[0]; vals = args[1:]
    if not n_clicks: return ""
    try:
        inp = np.array([
            float(vals[num_cols[:12].index(f)]) if f in num_cols[:12]
            and num_cols[:12].index(f)<len(vals)
            and vals[num_cols[:12].index(f)] is not None
            else float(df[f].mean()) if f in df.columns else 0.0
            for f in selected_features
        ])
        proba = best_model.predict_proba(inp.reshape(1,-1))[0]
        pred  = class_names[np.argmax(proba)]
        conf  = float(np.max(proba))
        color = clr_map.get(str(pred),"#58A6FF")

        if hasattr(best_model,"feature_importances_"): imps2=best_model.feature_importances_
        elif hasattr(best_model,"coef_"): imps2=np.abs(best_model.coef_).mean(axis=0);imps2/=(imps2.max()+1e-9)
        else: imps2=np.ones(len(selected_features))
        kpis   = sorted(zip(selected_features,imps2),key=lambda x:x[1],reverse=True)[:6]
        max_imp= kpis[0][1] if kpis else 1

        return html.Div([
            html.Div([
                html.Div("Predicted Class",style={"fontSize":"11px","color":PAL["sub"]}),
                html.Div(str(pred),style={"fontSize":"30px","fontWeight":"800","color":color}),
                html.Div(f"Confidence: {conf*100:.1f}%",style={"color":PAL["sub"],"fontSize":"13px"}),
            ],style={"background":color+"18","border":f"1px solid {color}44",
                     "borderLeft":f"5px solid {color}","borderRadius":"10px",
                     "padding":"14px","marginBottom":"14px"}),
            html.Div("Top KPI Recommendations",style={"fontWeight":"600","marginBottom":"10px"}),
            *[html.Div([
                html.Div(f"#{i+1}  {f.replace('_',' ').title()}",
                         style={"fontWeight":"600","fontSize":"13px"}),
                html.Div(style={"height":"5px","borderRadius":"3px","marginTop":"4px",
                                "background":f"linear-gradient(90deg,{color},{color}44)",
                                "width":f"{v/max_imp*100:.0f}%"}),
                html.Div(f"Score: {v:.4f}",style={"fontSize":"11px","color":PAL["sub"],"marginTop":"2px"}),
            ],style={"background":"#0D1117","border":f"1px solid {PAL['border']}",
                     "borderRadius":"8px","padding":"10px 14px","marginBottom":"8px"})
            for i,(f,v) in enumerate(kpis)],
        ])
    except Exception as e:
        return html.Div(f"Error: {e}",style={"color":"#F78166"})

print("✓ Callbacks registered")
print()
print("="*55)
print("  🚀 LAUNCHING DASHBOARD at http://127.0.0.1:8050")
print("  Open that URL in your browser now!")
print("  Press ■ Stop in Jupyter to shut it down.")
print("="*55)
```
