```python
import pandas as pd
import numpy as np
df = pd.read_csv("telecom_customer_churn_feature_engineering.csv")
```

```python
df[["monthly_income","monthly_bill"]].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual style for the plots
sns.set(style="whitegrid")

# Initialize the figure size
plt.figure(figsize=(10, 4))

# First Subplot: Monthly Income Distribution
plt.subplot(1, 2, 1)
sns.histplot(df["monthly_income"], kde=True, bins=30)
plt.title("Monthly Income Distribution")

# Second Subplot: Monthly Bill Distribution
plt.subplot(1, 2, 2)
sns.histplot(df["monthly_bill"], kde=True, bins=30)
plt.title("Monthly Bill Distribution")

# Adjust layout to prevent overlapping and display
plt.tight_layout()
plt.show()

# Observations noted in the image:
# Max >> Mean
# Strong skew present
# Ideal candidate for transformation
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- Safe Log Transformation ---
# np.log1p computes log(1 + x), which is safer for data containing zeros
df["income_log"] = np.log1p(df["monthly_income"])

# Preview the original and transformed columns
df[["monthly_income", "income_log"]].head()

# --- Visualization ---
sns.set(style="whitegrid")
plt.figure(figsize=(10, 4))

# Subplot 1: Original Scale
plt.subplot(1, 2, 1)
sns.histplot(df["monthly_income"], kde=True, bins=30)
plt.title("Monthly Income (Original Scale)")

# Subplot 2: Log Transformed Scale
plt.subplot(1, 2, 2)
sns.histplot(df["income_log"], kde=True, bins=30)
plt.title("Monthly Income (Log Transformed)")

plt.tight_layout()
plt.show()
```
