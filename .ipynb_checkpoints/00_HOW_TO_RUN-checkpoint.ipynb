## ✅ Step-by-Step Instructions

### Step 1 — Install dependencies


````
!pip install pandas numpy scikit-learn xgboost matplotlib plotly dash dash-bootstrap-components joblib scipy
print("✓ All packages installed")
````

### Step 2 — Set up your dataset
Place your CSV file in the `data/` folder.  
Then update `DATASET_PATH` and `TARGET_COL` in every notebook.

### Step 3 — Run notebooks IN ORDER

| Notebook | What it does |
|---|---|
| `00_preprocessing.ipynb` | Cleans, encodes, scales, selects features, splits data |
| `01_eda.ipynb` | Produces all EDA charts (Fig. 2 in paper) |
| `02_model_training.ipynb` | Trains RF, XGBoost, SVM, LR with GridSearchCV |
| `03_evaluation.ipynb` | All metrics, McNemar test, ROC curves, confusion matrices, feature importance, A/B simulation |
| `04_kpi_recommender.ipynb` | KPI prediction and role recommendation demo |
| `05_dashboard.ipynb` | Launches the full interactive web dashboard |

### Step 4 — Open your dashboard
After running `05_dashboard.ipynb`, open: **http://127.0.0.1:8050**

````
# Quick system check
import importlib, sys

packages = ["pandas","numpy","sklearn","xgboost","matplotlib",
            "plotly","dash","dash_bootstrap_components","joblib","scipy"]
all_ok = True
for pkg in packages:
    try:
        m = importlib.import_module(pkg)
        v = getattr(m,"__version__","ok")
        print(f"  ✓  {pkg:<30} {v}")
    except ImportError:
        print(f"  ✗  {pkg:<30} NOT INSTALLED — run: pip install {pkg}")
        all_ok = False

print()
if all_ok:
    print("✅  All packages ready. Open 00_preprocessing.ipynb to begin.")
else:
    print("❌  Some packages missing. Run the install cell above first.")
````

## 🎨 Dashboard Features
- **Dark glassmorphism** design with vibrant accent colors  
- **5 interactive tabs**: Overview, Analytics, ML Results, Features, Predict  
- **Live filters**: filter by class, switch scatter axes  
- **KPI gauge cards** for top numeric features  
- **Role prediction panel**: enter values → instant prediction  
- **Works with ANY CSV** — no code changes needed
