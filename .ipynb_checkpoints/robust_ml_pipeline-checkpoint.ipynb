# 🛡️ Robust Production-Ready ML Pipeline
**Works safely on ANY dataset — classification or regression**

---
## ✅ What this pipeline handles automatically:
| Problem | Solution |
|---|---|
| Feature mismatch train/test | Single `Pipeline` + `ColumnTransformer` — fit once, transform everywhere |
| Wrong task type detection | Auto-detects classification vs regression from target column |
| Missing values | `SimpleImputer` inside pipeline — never causes shape errors |
| Non-numeric columns | `OneHotEncoder` with `handle_unknown='ignore'` — no unseen category crashes |
| StratifiedKFold on tiny/imbalanced classes | Auto-switches to `KFold` when stratification is unsafe |
| Single-class target | Detected and aborted with clear error message before training |
| Empty dataset after cleaning | Detected immediately with helpful message |
| Regression mistaken as classification | Automatic detection by unique value count + dtype |
| Prediction shape errors | Prediction uses the same fitted pipeline — guaranteed match |
| Model saving/loading | `joblib` saves full pipeline including preprocessor |

## 🚀 How to Run:
1. Set `DATASET_PATH` and `TARGET_COL` in **Cell 2**
2. Click **Kernel → Restart & Run All**
3. Done — results appear automatically

## 📦 CELL 3 — Import All Libraries

```python
import warnings
warnings.filterwarnings("ignore")

import numpy  as np
import pandas as pd
import joblib, time, json
from collections import Counter

# sklearn — core
from sklearn.pipeline          import Pipeline
from sklearn.compose           import ColumnTransformer
from sklearn.preprocessing     import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.impute             import SimpleImputer
from sklearn.model_selection   import (
    cross_val_score, cross_validate,
    StratifiedKFold, KFold,
    train_test_split
)
from sklearn.metrics import (
    # classification
    accuracy_score, f1_score, precision_score, recall_score,
    roc_auc_score, classification_report, confusion_matrix,
    # regression
    mean_squared_error, mean_absolute_error, r2_score,
)
from sklearn.ensemble          import RandomForestClassifier, RandomForestRegressor
from sklearn.svm               import SVC, SVR
from sklearn.linear_model      import LogisticRegression, Ridge
from xgboost                   import XGBClassifier, XGBRegressor

import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

print("✓ All imports successful")
```

## 🔍 CELL 4 — Load & Validate Dataset

```python
# ── Load ─────────────────────────────────────────────────────────
if not os.path.isfile(DATASET_PATH):
    raise FileNotFoundError(
        f"\n❌ File not found: '{DATASET_PATH}'\n"
        f"   Please check the path in Cell 2."
    )

df_raw = pd.read_csv(DATASET_PATH)
print(f"✓ Loaded  : {df_raw.shape[0]:,} rows × {df_raw.shape[1]} columns")

# ── Validate target column ────────────────────────────────────────
if TARGET_COL not in df_raw.columns:
    raise ValueError(
        f"\n❌ Target column '{TARGET_COL}' not found.\n"
        f"   Available columns: {list(df_raw.columns)}"
    )

# ── Drop rows where target is missing ────────────────────────────
before = len(df_raw)
df_raw = df_raw.dropna(subset=[TARGET_COL]).reset_index(drop=True)
dropped = before - len(df_raw)
if dropped:
    print(f"  ⚠ Dropped {dropped} rows with missing target values")

# ── Guard: empty dataset ──────────────────────────────────────────
if len(df_raw) == 0:
    raise ValueError("❌ Dataset is empty after removing missing targets. Check your CSV.")

if len(df_raw) < 10:
    raise ValueError(f"❌ Dataset has only {len(df_raw)} rows — too small to train reliably.")

print(f"  Rows used : {len(df_raw):,}")
print(f"  Missing   : {df_raw.isnull().sum().sum():,} total values across all columns")
print()
print("Dataset preview:")
df_raw.head(3)
```

---
## 🤖 CELL 5 — Auto-Detect Task Type (Classification vs Regression)

```python
def detect_task_type(series, name="target"):
    """
    Robustly determine whether a target column is classification or regression.

    Rules (in order):
      1. Object / category dtype              → classification
      2. Boolean dtype                        → classification
      3. Unique values ≤ 20 AND all integers  → classification
      4. Unique ratio < 5% of rows            → classification
      5. Otherwise                            → regression
    """
    s        = series.dropna()
    n        = len(s)
    n_unique = s.nunique()
    dtype    = s.dtype

    reasons = []

    if dtype == object or str(dtype) == "category":
        reasons.append("string/category dtype")
        task = "classification"

    elif dtype == bool:
        reasons.append("boolean dtype")
        task = "classification"

    elif n_unique <= 2:
        reasons.append(f"only {n_unique} unique value(s)")
        task = "classification"

    elif n_unique <= 20 and pd.api.types.is_integer_dtype(dtype):
        reasons.append(f"{n_unique} unique integers (≤ 20 threshold)")
        task = "classification"

    elif (n_unique / n) < 0.05:
        reasons.append(f"unique ratio = {n_unique/n:.2%} < 5%")
        task = "classification"

    else:
        reasons.append(f"{n_unique} unique values ({n_unique/n:.1%} of rows), float/int dtype")
        task = "regression"

    return task, reasons


TASK, reasons = detect_task_type(df_raw[TARGET_COL], TARGET_COL)

print(f"{'='*55}")
print(f"  🎯 TASK DETECTED: {TASK.upper()}")
print(f"{'='*55}")
print(f"  Target column : '{TARGET_COL}'")
print(f"  Dtype         : {df_raw[TARGET_COL].dtype}")
print(f"  Unique values : {df_raw[TARGET_COL].nunique()}")
print(f"  Detection reasons:")
for r in reasons:
    print(f"    • {r}")

if TASK == "classification":
    vc = df_raw[TARGET_COL].value_counts()
    print(f"\n  Class distribution:")
    print(vc.to_string())

    # Guard: single class
    if df_raw[TARGET_COL].nunique() < 2:
        raise ValueError(
            "❌ Target has only 1 unique class — cannot train a classifier.\n"
            "   Check your target column or dataset filtering."
        )
```

---
## 🔧 CELL 6 — Prepare Features & Encode Target

```python
# ── Separate features and target ──────────────────────────────────
X_raw = df_raw.drop(columns=[TARGET_COL])
y_raw = df_raw[TARGET_COL].copy()

# ── Encode classification target safely ───────────────────────────
le = None
if TASK == "classification":
    le = LabelEncoder()
    y  = le.fit_transform(y_raw.astype(str))
    print(f"✓ Target encoded | Classes: {list(le.classes_)}")
else:
    y = y_raw.values.astype(float)
    print(f"✓ Target kept as float | Range: {y.min():.4f} – {y.max():.4f}")

# ── Identify column types ─────────────────────────────────────────
numeric_cols     = X_raw.select_dtypes(include=["number"]).columns.tolist()
categorical_cols = X_raw.select_dtypes(exclude=["number"]).columns.tolist()

# Drop high-cardinality categorical columns (> 50 unique values)
# They cause memory explosions in OneHotEncoder
high_card = [c for c in categorical_cols if X_raw[c].nunique() > 50]
if high_card:
    print(f"  ⚠ Dropping {len(high_card)} high-cardinality column(s) "
          f"(> 50 unique vals): {high_card}")
    X_raw = X_raw.drop(columns=high_card)
    categorical_cols = [c for c in categorical_cols if c not in high_card]

print(f"\n✓ Feature matrix  : {X_raw.shape[0]:,} rows × {X_raw.shape[1]} columns")
print(f"  Numeric cols    : {len(numeric_cols)}  → {numeric_cols[:6]}{'...' if len(numeric_cols)>6 else ''}")
print(f"  Categorical cols: {len(categorical_cols)}  → {categorical_cols[:6]}{'...' if len(categorical_cols)>6 else ''}")
print(f"  Missing values  : {X_raw.isnull().sum().sum():,}")
```

---
## 🏗️ CELL 7 — Build ColumnTransformer Preprocessing Pipeline

```python
# ─────────────────────────────────────────────────────────────────
# KEY DESIGN PRINCIPLE:
# Everything is inside ONE sklearn Pipeline.
# fit() is called ONLY on training data.
# transform() is called on test data — guaranteed same columns.
# No manual encoding that could cause train/test mismatch.
# ─────────────────────────────────────────────────────────────────

# ── Numeric transformer ───────────────────────────────────────────
# Impute with median (robust to outliers), then StandardScale
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler",  StandardScaler()),
])

# ── Categorical transformer ───────────────────────────────────────
# Impute with most frequent, then OneHotEncode
# handle_unknown='ignore' → unseen categories become all-zeros (no crash)
# sparse_output=False     → avoids sparse matrix issues with some models
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
])

# ── Combine into ColumnTransformer ───────────────────────────────
transformers = []
if numeric_cols:
    transformers.append(("num", numeric_transformer, numeric_cols))
if categorical_cols:
    transformers.append(("cat", categorical_transformer, categorical_cols))

if not transformers:
    raise ValueError("❌ No valid feature columns found after cleaning.")

preprocessor = ColumnTransformer(
    transformers=transformers,
    remainder="drop",          # safely drops any unexpected columns
    verbose_feature_names_out=False,
)

print("✓ Preprocessing pipeline built")
print(f"  Numeric    : {len(numeric_cols)} columns → median impute + StandardScaler")
print(f"  Categorical: {len(categorical_cols)} columns → mode impute + OneHotEncoder(handle_unknown='ignore')")
print(f"  Unknown categories at prediction time: silently ignored (no crash)")
```

---
## 🔀 CELL 8 — Train / Test Split

```python
# ── Safe stratified split ─────────────────────────────────────────
# Stratification fails when a class has < 2 members.
# We check first and fall back to random split safely.

TEST_SIZE   = 0.20
RANDOM_SEED = 42

def safe_train_test_split(X, y, task, test_size=0.20, random_state=42):
    """
    Stratified split for classification when safe.
    Falls back to random split when:
      - regression task
      - any class has fewer than 2 samples
      - dataset is very small
    """
    if task == "regression":
        X_tr, X_te, y_tr, y_te = train_test_split(
            X, y, test_size=test_size, random_state=random_state)
        return X_tr, X_te, y_tr, y_te, "random"

    counts    = Counter(y)
    min_count = min(counts.values())

    if min_count < 2:
        print(f"  ⚠ Class with only {min_count} sample(s) found — using random split")
        X_tr, X_te, y_tr, y_te = train_test_split(
            X, y, test_size=test_size, random_state=random_state)
        return X_tr, X_te, y_tr, y_te, "random"

    X_tr, X_te, y_tr, y_te = train_test_split(
        X, y, test_size=test_size, stratify=y, random_state=random_state)
    return X_tr, X_te, y_tr, y_te, "stratified"


X_train, X_test, y_train, y_test, split_type = safe_train_test_split(
    X_raw.values if not categorical_cols else X_raw,
    y, TASK, TEST_SIZE, RANDOM_SEED
)

# Keep as DataFrame to preserve column names for ColumnTransformer
if not isinstance(X_train, pd.DataFrame):
    X_train = pd.DataFrame(X_train, columns=X_raw.columns)
    X_test  = pd.DataFrame(X_test,  columns=X_raw.columns)

print(f"✓ Split type : {split_type}")
print(f"  Train set  : {len(X_train):,} rows")
print(f"  Test set   : {len(X_test):,}  rows")
if TASK == "classification":
    print(f"  Train class distribution: {dict(Counter(y_train))}")
    print(f"  Test  class distribution: {dict(Counter(y_test))}")
```

---
## 🤖 CELL 9 — Define Models

```python
def get_models(task):
    """
    Return appropriate models based on detected task type.
    All models wrapped inside Pipeline with the shared preprocessor.
    """
    n_samples  = len(X_train)
    n_features = X_raw.shape[1]

    if task == "classification":
        base_models = {
            "Random Forest": RandomForestClassifier(
                n_estimators=200,
                max_depth=None,
                min_samples_split=2,
                random_state=RANDOM_SEED,
                n_jobs=-1,
            ),
            "XGBoost": XGBClassifier(
                n_estimators=200,
                max_depth=6,
                learning_rate=0.1,
                random_state=RANDOM_SEED,
                eval_metric="logloss",
                verbosity=0,
                use_label_encoder=False,
                n_jobs=-1,
            ),
            "SVM": SVC(
                C=10,
                kernel="rbf",
                gamma="scale",
                probability=True,   # needed for predict_proba
                random_state=RANDOM_SEED,
            ),
            "Logistic Regression": LogisticRegression(
                C=1.0,
                solver="lbfgs",
                max_iter=1000,
                random_state=RANDOM_SEED,
                n_jobs=-1,
            ),
        }
    else:
        base_models = {
            "Random Forest": RandomForestRegressor(
                n_estimators=200,
                max_depth=None,
                random_state=RANDOM_SEED,
                n_jobs=-1,
            ),
            "XGBoost": XGBRegressor(
                n_estimators=200,
                max_depth=6,
                learning_rate=0.1,
                random_state=RANDOM_SEED,
                verbosity=0,
                n_jobs=-1,
            ),
            "SVM": SVR(
                C=10,
                kernel="rbf",
                gamma="scale",
            ),
            "Ridge Regression": Ridge(
                alpha=1.0,
                random_state=RANDOM_SEED,
            ),
        }

    # Wrap every model with the shared preprocessor inside a Pipeline
    # This guarantees: fit preprocessor on train → apply same transform to test
    pipelines = {
        name: Pipeline(steps=[
            ("preprocessor", preprocessor),
            ("model",        mdl),
        ])
        for name, mdl in base_models.items()
    }

    return pipelines


PIPELINES = get_models(TASK)

print(f"✓ Task type : {TASK.upper()}")
print(f"  Models ready:")
for name in PIPELINES:
    print(f"    • {name}")
print()
print("  Each model is wrapped in Pipeline([preprocessor, model])")
print("  → preprocessor is fit ONLY on training data")
print("  → test data is transformed using the same fitted preprocessor")
print("  → NO feature mismatch possible")
```

---
## 🔁 CELL 10 — Safe Cross-Validation Strategy

```python
def get_cv_strategy(task, y_train, n_splits=5):
    """
    Automatically choose the safest CV strategy:
    - StratifiedKFold for classification when all classes have enough samples
    - KFold otherwise (regression, tiny classes, very small datasets)

    Also adjusts n_splits down if dataset is very small.
    """
    n = len(y_train)

    # Reduce folds if dataset is small
    max_safe_splits = n // 5   # need at least 5 samples per fold
    n_splits        = min(n_splits, max(2, max_safe_splits))

    if task == "regression":
        cv = KFold(n_splits=n_splits, shuffle=True, random_state=RANDOM_SEED)
        return cv, n_splits, "KFold (regression)"

    # Classification — check if stratification is safe
    counts    = Counter(y_train)
    min_count = min(counts.values())

    if min_count < n_splits:
        # Cannot stratify: some class has fewer samples than folds
        cv = KFold(n_splits=min(n_splits, min_count), shuffle=True, random_state=RANDOM_SEED)
        actual = min(n_splits, min_count)
        return cv, actual, f"KFold (class too small: min_count={min_count})"

    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=RANDOM_SEED)
    return cv, n_splits, "StratifiedKFold"


CV, N_SPLITS, CV_REASON = get_cv_strategy(TASK, y_train, n_splits=5)

print(f"✓ CV strategy : {CV_REASON}")
print(f"  Folds       : {N_SPLITS}")
print(f"  Training rows per fold: ~{len(y_train)*(N_SPLITS-1)//N_SPLITS}")
print(f"  Validation rows per fold: ~{len(y_train)//N_SPLITS}")
```

---
## 🏋️ CELL 11 — Train All Models with Cross-Validation

```python
SCORING = {
    "classification": "f1_weighted",
    "regression":     "r2",
}[TASK]

cv_results  = {}
all_results = {}
total_start = time.time()

print(f"{'='*60}")
print(f"  TRAINING — {TASK.upper()} | CV={CV_REASON} | Scoring={SCORING}")
print(f"{'='*60}")

for name, pipeline in PIPELINES.items():
    print(f"\n  ▶ {name} ...")
    t0 = time.time()

    try:
        scores = cross_val_score(
            pipeline, X_train, y_train,
            cv=CV, scoring=SCORING, n_jobs=-1,
        )
        cv_results[name] = scores
        print(f"    CV {SCORING}: {scores.mean():.4f} ± {scores.std():.4f}  "
              f"[{scores.min():.4f} – {scores.max():.4f}]  "
              f"| {time.time()-t0:.1f}s")

    except Exception as e:
        print(f"    ⚠ Skipped — {e}")
        cv_results[name] = np.array([np.nan])

print(f"\n{'='*60}")
print(f"  Total training time: {time.time()-total_start:.1f}s")
```

---
## 🏆 CELL 12 — Select Best Model & Evaluate on Test Set

```python
# ── Pick best model by mean CV score ─────────────────────────────
valid = {k: v for k, v in cv_results.items() if not np.isnan(v.mean())}
if not valid:
    raise RuntimeError("❌ All models failed cross-validation. Check your dataset.")

best_name = max(valid, key=lambda k: valid[k].mean())
best_pipe = PIPELINES[best_name]
print(f"★  Best CV model: {best_name}  ({SCORING}={valid[best_name].mean():.4f})")

# ── Fit best model on FULL training set ──────────────────────────
print(f"\n  Fitting {best_name} on full training set ...")
t0 = time.time()
best_pipe.fit(X_train, y_train)
print(f"  ✓ Fitted in {time.time()-t0:.1f}s")

# ── Also fit all other valid models for comparison ────────────────
all_test_results = {}
for name, pipeline in PIPELINES.items():
    if name not in valid: continue
    if name != best_name:
        pipeline.fit(X_train, y_train)

    y_pred = pipeline.predict(X_test)

    if TASK == "classification":
        acc  = accuracy_score(y_test, y_pred)
        f1   = f1_score(y_test, y_pred, average="weighted", zero_division=0)
        prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
        rec  = recall_score(y_test, y_pred, average="weighted", zero_division=0)
        try:
            y_proba = pipeline.predict_proba(X_test)
            if y_proba.shape[1] == 2:
                auc = roc_auc_score(y_test, y_proba[:,1])
            else:
                auc = roc_auc_score(y_test, y_proba, multi_class="ovr", average="weighted")
        except Exception:
            auc = float("nan")
        all_test_results[name] = {
            "Accuracy":  round(acc,  4),
            "F1":        round(f1,   4),
            "Precision": round(prec, 4),
            "Recall":    round(rec,  4),
            "ROC-AUC":   round(auc,  4),
            "CV_mean":   round(valid[name].mean(), 4),
            "CV_std":    round(valid[name].std(),  4),
        }
    else:
        mse  = mean_squared_error(y_test, y_pred)
        mae  = mean_absolute_error(y_test, y_pred)
        r2   = r2_score(y_test, y_pred)
        rmse = np.sqrt(mse)
        all_test_results[name] = {
            "R²":      round(r2,   4),
            "RMSE":    round(rmse, 4),
            "MAE":     round(mae,  4),
            "MSE":     round(mse,  4),
            "CV_mean": round(valid[name].mean(), 4),
            "CV_std":  round(valid[name].std(),  4),
        }

df_results = pd.DataFrame(all_test_results).T
df_results.index.name = "Model"
df_results.to_csv("outputs/model_results.csv")

print(f"\n{'='*60}")
print(f"  TEST SET RESULTS — {TASK.upper()}")
print(f"{'='*60}")
print(df_results.to_string())
print(f"\n  ★  Best model: {best_name}")
```

## 📊 CELL 13 — Classification Report & Confusion Matrix

```python
y_best_pred = best_pipe.predict(X_test)

if TASK == "classification":
    # ── Classification report ─────────────────────────────────────
    target_names = list(le.classes_) if le else [str(c) for c in np.unique(y_test)]
    report = classification_report(y_test, y_best_pred,
                                   target_names=target_names, zero_division=0)
    print(f"CLASSIFICATION REPORT — {best_name}")
    print("="*55)
    print(report)

    # ── Confusion matrix chart ────────────────────────────────────
    cm = confusion_matrix(y_test, y_best_pred)
    fig, ax = plt.subplots(figsize=(max(5, len(target_names)*1.5),
                                    max(4, len(target_names)*1.2)))
    im = ax.imshow(cm, cmap="Blues")
    ax.set_xticks(range(len(target_names)))
    ax.set_yticks(range(len(target_names)))
    ax.set_xticklabels(target_names, rotation=45, ha="right", fontsize=9)
    ax.set_yticklabels(target_names, fontsize=9)
    for i in range(len(target_names)):
        for j in range(len(target_names)):
            ax.text(j, i, str(cm[i,j]), ha="center", va="center",
                    fontsize=11, fontweight="bold",
                    color="white" if cm[i,j] > cm.max()*0.5 else "black")
    plt.colorbar(im, ax=ax)
    ax.set_xlabel("Predicted", fontsize=11)
    ax.set_ylabel("Actual",    fontsize=11)
    ax.set_title(f"Confusion Matrix — {best_name}", fontsize=13, fontweight="bold")
    plt.tight_layout()
    plt.savefig("outputs/confusion_matrix.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("✓ Saved: outputs/confusion_matrix.png")

else:
    # ── Actual vs Predicted chart ─────────────────────────────────
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Scatter: actual vs predicted
    axes[0].scatter(y_test, y_best_pred, alpha=0.5, color="#58A6FF", s=20)
    mn = min(y_test.min(), y_best_pred.min())
    mx = max(y_test.max(), y_best_pred.max())
    axes[0].plot([mn,mx],[mn,mx],"r--",linewidth=1.5,label="Perfect prediction")
    axes[0].set_xlabel("Actual",    fontsize=11)
    axes[0].set_ylabel("Predicted", fontsize=11)
    axes[0].set_title(f"Actual vs Predicted — {best_name}", fontsize=12, fontweight="bold")
    axes[0].legend(fontsize=9)
    axes[0].spines[["top","right"]].set_visible(False)

    # Residuals histogram
    residuals = y_test - y_best_pred
    axes[1].hist(residuals, bins=30, color="#3FB950", edgecolor="white", alpha=0.85)
    axes[1].axvline(0, color="red", linestyle="--", linewidth=1.5)
    axes[1].set_xlabel("Residual (Actual − Predicted)", fontsize=11)
    axes[1].set_ylabel("Count",  fontsize=11)
    axes[1].set_title("Residual Distribution", fontsize=12, fontweight="bold")
    axes[1].spines[["top","right"]].set_visible(False)

    plt.tight_layout()
    plt.savefig("outputs/regression_evaluation.png", dpi=150, bbox_inches="tight")
    plt.show()

    r2_val   = r2_score(y_test, y_best_pred)
    rmse_val = np.sqrt(mean_squared_error(y_test, y_best_pred))
    mae_val  = mean_absolute_error(y_test, y_best_pred)
    print(f"  R²   = {r2_val:.4f}")
    print(f"  RMSE = {rmse_val:.4f}")
    print(f"  MAE  = {mae_val:.4f}")
    print("✓ Saved: outputs/regression_evaluation.png")
```

## 🔍 CELL 15 — Feature Importance

```python
# Extract feature names after ColumnTransformer
try:
    feature_names_out = best_pipe.named_steps["preprocessor"].get_feature_names_out()
except Exception:
    feature_names_out = None

# Get importances from the model step
model_step = best_pipe.named_steps["model"]

if hasattr(model_step, "feature_importances_"):
    imps = model_step.feature_importances_
    imp_type = "feature_importances_"
elif hasattr(model_step, "coef_"):
    coef = model_step.coef_
    if coef.ndim > 1:
        imps = np.abs(coef).mean(axis=0)
    else:
        imps = np.abs(coef)
    imp_type = "coef_ (absolute mean)"
else:
    imps = None
    imp_type = "not available"

if imps is not None and feature_names_out is not None and len(imps)==len(feature_names_out):
    top_n   = min(20, len(feature_names_out))
    idx     = np.argsort(imps)[-top_n:]
    feats   = feature_names_out[idx]
    vals    = imps[idx]
    norm    = vals / (vals.max()+1e-9)
    bar_c   = [f"hsl({int(200+v*140)},70%,{int(40+v*30)}%)" for v in norm]

    fig, ax = plt.subplots(figsize=(10, max(5, top_n*0.4)))
    bars = ax.barh(feats, vals, color=bar_c, edgecolor="white", height=0.65)
    for bar, v in zip(bars, vals):
        ax.text(bar.get_width()+vals.max()*0.01, bar.get_y()+bar.get_height()/2,
                f"{v:.4f}", va="center", fontsize=9, fontweight="bold")
    ax.set_xlabel(f"Importance ({imp_type})", fontsize=11)
    ax.set_title(f"Top-{top_n} Feature Importances — {best_name}", fontsize=12, fontweight="bold")
    ax.spines[["top","right"]].set_visible(False)
    ax.xaxis.grid(True, linestyle="--", alpha=0.4); ax.set_axisbelow(True)
    plt.tight_layout()
    plt.savefig("outputs/feature_importance.png", dpi=150, bbox_inches="tight")
    plt.show()

    df_imp = pd.DataFrame({"feature":feature_names_out,"importance":imps})                .sort_values("importance",ascending=False).reset_index(drop=True)
    df_imp.to_csv("outputs/feature_importance.csv", index=False)
    print(f"✓ Saved: outputs/feature_importance.png")
    print(f"✓ Saved: outputs/feature_importance.csv")
    print(f"\nTop 10:")
    print(df_imp.head(10).to_string(index=False))
else:
    print(f"  ⚠ Feature importance not available for {best_name} ({imp_type})")
```

## 🧪 CELL 17 — Safe Prediction on New Data

```python
def safe_predict(pipeline, new_data, task, label_encoder=None):
    """
    Safely predict on new raw data.
    - Accepts DataFrame with same column names as training data
    - Handles missing columns by filling with NaN (imputer handles them)
    - Handles extra columns by ignoring them
    - Returns predictions (and class names for classification)
    """
    # Ensure DataFrame
    if not isinstance(new_data, pd.DataFrame):
        raise TypeError("new_data must be a pandas DataFrame")

    # Align columns: keep only training columns, fill missing with NaN
    expected_cols = meta["feature_cols"]
    extra   = [c for c in new_data.columns if c not in expected_cols]
    missing = [c for c in expected_cols    if c not in new_data.columns]

    if extra:
        print(f"  ⚠ Ignoring {len(extra)} extra column(s): {extra}")
    if missing:
        print(f"  ⚠ Filling {len(missing)} missing column(s) with NaN: {missing}")
        for col in missing:
            new_data[col] = np.nan

    new_data = new_data[expected_cols].copy()

    # Predict
    y_pred_enc = pipeline.predict(new_data)

    # Decode classification labels
    if task == "classification" and label_encoder is not None:
        y_pred_labels = label_encoder.inverse_transform(y_pred_enc)
        return y_pred_enc, y_pred_labels
    return y_pred_enc, None


# ── Demo: predict on first 5 rows of test set ────────────────────
print("Demo — predicting on first 5 test rows:")
sample = X_test.head(5).copy()
pred_enc, pred_labels = safe_predict(best_pipe, sample, TASK, le)

if TASK == "classification":
    actual_labels = le.inverse_transform(y_test[:5]) if le else y_test[:5]
    for i, (actual, pred) in enumerate(zip(actual_labels, pred_labels)):
        status = "✓" if actual==pred else "✗"
        print(f"  Row {i+1}: Actual={actual:<15}  Predicted={pred:<15}  {status}")
else:
    actual_vals = y_test[:5]
    for i, (actual, pred) in enumerate(zip(actual_vals, pred_enc)):
        err = abs(actual - pred)
        print(f"  Row {i+1}: Actual={actual:.4f}  Predicted={pred:.4f}  |Error|={err:.4f}")
```
