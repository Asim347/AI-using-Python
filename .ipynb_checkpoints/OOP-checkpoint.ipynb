```python
# Instance varibales & Methods
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def display_info(self):
        print(f"Student Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Grade: {self.grade}")

    def is_eligible(self):
        if self.age >= 15:
            print(self.name, "is eligible for admission.")
        else:
            print(self.name, "is not eligible for admission.")
```

```python
student1 = Student("asim", 15, "C")
student3 = Student("ali", 19, "A")
```

```python
print(student1.name)
print(student3.name)
```

```python
student1.display_info();
student1.is_eligible();
```

```python
student3.display_info();
student3.is_eligible();
```

```python
class Student:
    def __init__(self, name, grade):
        self.__name = name   # Private attribute
        self.grade = grade    # Public attribute

    # Getter method to access private data
    def get_name(self):
        return self.__name

# Creating the instance
std = Student("ali", "A")

# This will now work correctly using the getter
print(std.get_name()) 

# This would still cause an AttributeError
# print(std.__name)
```

```python
class Student:
    @property
    def grade(self):
        return self.__grade

    def __init__(self, name, grade):
        self.__grade = grade

    def get_grade(self):
        return self.__grade

    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print("Invalid grade!")

# Testing the code
student = Student("Ali", 90)
student.set_grade(95)
print(student.get_grade())
```

```python
# Inheritance
# Parent Class
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def study(self):
        print(f"{self.name} is studying.")

# Child Class (Inherits from Student)
class GraduateStudent(Student):
    def __init__(self, name, age, thesis_topic):
        # super() calls the __init__ of the Parent class
        super().__init__(name, age) 
        self.thesis_topic = thesis_topic

    def research(self):
        print(f"{self.name} is researching: {self.thesis_topic}")

# --- Testing the Inheritance ---

# 1. Regular Student
std1 = Student("Asim", 20)
std1.study()

# 2. Graduate Student
grad1 = GraduateStudent("Sara", 25, "AI in Healthcare")
grad1.study()    # Inherited from Student
grad1.research() # Unique to GraduateStudent
```

```python
# Parent Class
class Student:
    def study(self):
        print("Student is reading books.")

# Child Class
class MedicalStudent(Student):
    # This OVERRIDES the study method above
    def study(self):
        print("Medical Student is practicing in the lab and studying anatomy.")

# --- Testing Overriding ---

std = Student()
med_std = MedicalStudent()

std.study()      # Calls the Parent version
med_std.study()  # Calls the Overridden Child version
```

```python
# Ploymorphism
# Parent Class
class Animal:
    def make_sound(self):
        # A generic method that will be replaced by children
        pass

# Child Class 1
class Dog(Animal):
    def make_sound(self):
        print("Woof! Woof!")

# Child Class 2
class Cat(Animal):
    def make_sound(self):
        print("Meow! Meow!")

# --- Polymorphism in Action ---

# We can put different animals in the same list
animals = [Dog(), Cat()]

# We call the SAME method name on both, but get DIFFERENT results
for animal in animals:
    animal.make_sound()
```
