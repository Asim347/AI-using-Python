```python
# Demonstration of Naive Bayes and the feature independence assumption

import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# -----------------------------------------------------------------
# 1) Load dataset
# -----------------------------------------------------------------
data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Number of features:", X_train.shape[1])

# -----------------------------------------------------------------
# 2) Train Naive Bayes classifier
# -----------------------------------------------------------------
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
# -----------------------------------------------------------------
# # 3) Evaluate performance
# -----------------------------------------------------------------
y_pred = nb.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print(f"Naive Bayes Accuracy: {acc:.4f}")

# -----------------------------------------------------------------
# # 4) Show learned probabilities (for intuition)
# -----------------------------------------------------------------
print("\nClass prior probabilities P(y):")
print(nb.class_prior_)

print("\nFeature means per class (sample of first 5 features):")
print(nb.theta_[:, :5])

print("\nFeature variances per class (sample of first 5 features):")
print(nb.var_[:, :5])
```

```python
# Demonstration of three Naive Bayes models

from sklearn.datasets import load_breast_cancer, make_classification
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score

# -----------------------------------------------------------------
# # 1) GAUSSIAN NAIVE BAYES -> continuous data
# -----------------------------------------------------------------
X_g, y_g = load_breast_cancer(return_X_y=True)

X_tr, X_te, y_tr, y_te = train_test_split(
    X_g, y_g, test_size=0.25, random_state=42, stratify=y_g
)

g_nb = GaussianNB()
g_nb.fit(X_tr, y_tr)
acc_g = accuracy_score(y_te, g_nb.predict(X_te))
acc_g
```

```python
# -----------------------------------------------------------------
# # 2) MULTINOMIAL NAIVE BAYES -> text / count data
# -----------------------------------------------------------------
texts = [
    "buy cheap medicine for free",
    "win a free prize",
    "meeting schedule confirmed",
    "project deadline tomorrow. we must do this",
    "limited time offer to win prize",
    "let us discuss the report at your free schedule"
]
labels = [1, 1, 0, 0, 1, 0]  # 1 = spam, 0 = not spam

vectorizer = CountVectorizer()
X_text = vectorizer.fit_transform(texts)

X_tr, X_te, y_tr, y_te = train_test_split(
    X_text, labels, test_size=0.33, random_state=42
)

m_nb = MultinomialNB()
m_nb.fit(X_tr, y_tr)
acc_m = accuracy_score(y_te, m_nb.predict(X_te))
acc_m


# -----------------------------------------------------------------
# # 3) BERNOULLI NAIVE BAYES -> binary features
# -----------------------------------------------------------------
X_b, y_b = make_classification(
    n_samples=200, n_features=10, n_informative=5,
    n_redundant=0, random_state=42
)

# Convert to binary (0/1)
X_b = (X_b > 0).astype(int)

X_tr, X_te, y_tr, y_te = train_test_split(
    X_b, y_b, test_size=0.25, random_state=42
)

b_nb = BernoulliNB()
b_nb.fit(X_tr, y_tr)
acc_b = accuracy_score(y_te, b_nb.predict(X_te))


# -----------------------------------------------------------------
# # Results
# -----------------------------------------------------------------
print("\nModel performance:")
print(f"Gaussian NB accuracy    : {acc_g:.3f}")
print(f"Multinomial NB accuracy : {acc_m:.3f}")
```

```python
# ==========================================
# # 1. Import Required Libraries
# ==========================================
import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ==========================================
# # 2. Load Dataset
# ==========================================
# Breast Cancer dataset is a real-world medical dataset
data = load_breast_cancer()

# Features and target
X = data.data
y = data.target

# Convert to DataFrame for better understanding
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

print("Dataset Shape:", df.shape)
print("\nFirst 5 rows of dataset:\n")
print(df.head())
```

```python
# ==========================================
# # 3. Dataset Information
# ==========================================
print("\nTarget Classes:")
print(data.target_names)

print("\nClass Distribution:")
print(df['target'].value_counts())

# ==========================================
# # 4. Split Dataset into Training and Testing Sets
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

print("\nTraining samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
# ==========================================
# # 5. Create and Train Naive Bayes Model
# ==========================================
# GaussianNB is used for continuous numerical data
nb_model = GaussianNB()

# Train the model
nb_model.fit(X_train, y_train)

print("\nNaive Bayes model training completed.")
```

```python
# ==========================================
# # 6. Make Predictions
# ==========================================
y_pred = nb_model.predict(X_test)

print("\nFirst 10 Predictions:")
print(y_pred[:10])
```

```python
# ==========================================
# # 7. Model Evaluation
# ==========================================

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("\nModel Accuracy:", round(accuracy * 100, 2), "%")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Detailed classification metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# ==========================================
# # 8. Interpretation of Confusion Matrix
# ==========================================
print("\nConfusion Matrix Interpretation:")
print("True Negative (TN):", cm[0][0])
print("False Positive (FP):", cm[0][1])
print("False Negative (FN):", cm[1][0])
print("True Positive (TP):", cm[1][1])
```

```python
# ==========================================
# # 9. Predict on New Sample (Example)
# ==========================================
sample = X_test[0].reshape(1, -1)
prediction = nb_model.predict(sample)

print("\nPrediction for a new sample:", 
      data.target_names[prediction][0])
```
