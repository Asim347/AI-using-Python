```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "City" : ["Peshawar","lahore", "Quetta","Karachi"],
    "Segment" : ["A","B","C","D"],
    "Sales" : [200,300,400,500]

})
print(df)
```

```python
# one- hot encdoe
encoded = pd.get_dummies(df,columns = ["City","Segment" ], drop_first = False)
print(encoded)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# --- Part 1: Ordinal Encoding ---
# Creating a sample DataFrame
df2 = pd.DataFrame({
    "Size": ["Small", "Medium", "Large", "Medium", "Large"]
})

print("Original DataFrame:")
print(df2)
```

```python
# Initializing the OrdinalEncoder with a specific order
encoder = OrdinalEncoder(categories=[["Small", "Medium", "Large"]])

# Fitting and transforming the 'Size' column
df2["Size_encoded"] = encoder.fit_transform(df2[["Size"]])

print("\nDataFrame after Ordinal Encoding:")
print(df2)
```
