```python
import numpy as np

# Creating a 2D array (3x3 matrix)
arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])
```

```python
print(arr[0, 1])  # Output: 20 (row 0, col 1)
print(arr[2, 2])  # Output: 90 (row 2, col 2)

# Negative Indexing (counting from the end)
print(arr[-1, -2]) # Output: 80 (last row, 2nd last column)
print(arr[-3, -3]) # Output: 10 (first row, first column)
```

```python
# General form: arr[start:stop:step]
print(arr[0:2, 1:3]) # Output: rows 0-1, cols 1-2

# Striding: skipping elements
print(arr[::2, ::2]) # Output: every 2nd row & 2nd column
```

```python
# Slicing creates a "view" (editing the slice affects the original)
slice_arr = arr[0, :2] # view (not a copy)
slice_arr[0] = 999

print(arr[0])          # Original changes to [999, 20, 30]

# Using .copy() creates an independent object
copy_arr = arr[1].copy() # independent copy
copy_arr[0] = 555

print(arr[1])          # Original remains unchanged
```

```python
# Random generator with seed for reproducibility
rng = np.random.default_rng(seed = 43)
flips = rng.integers(0, 2, size = 100000)
pro_heads = np.mean(flips)
print("Estimated probability of heads : ", pro_heads)
```

```python
# Simulate 100k coin flips
rng = np.random.default_rng(seed=43)
flips = rng.integers(0, 2, size=100000) # 0 = tails, 1 = heads
prob_heads = np.mean(flips)
print("Estimated probability of heads:", prob_heads)

import time

# Python loop
start = time.time()
squares = [i**2 for i in range(1000000)]
end = time.time()
print("Loop time:", end - start)

# NumPy vectorized
start = time.time()
arr = np.arange(1000000)
squares_np = arr**2
end = time.time()
print("Vectorized time:", end - start)
```
