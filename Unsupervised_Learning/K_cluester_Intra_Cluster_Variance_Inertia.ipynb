```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import os
os.environ["OMP_NUM_THREADS"] = "1"
```

```python
X, _ = make_blobs(
    n_samples = 200,
    centers = 3,
    cluster_std = 1.0,
    random_state = 42
)
```

```python
k = 3
kmeans = KMeans(n_clusters = k, random_state = 42, n_init = 10)
labels = kmeans.fit_predict(X)
```

```python
plt.figure(figsize = (6,5))
plt.scatter(X[:,0], X[:,1], c = labels, cmap = "tab10" , alpha = 0.7)
plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s = 200, marker = "X", color = "black",
    label = "Centeroids"
)
plt.title("K-Means Clustering with k = 3")
plt.legend()
plt.show()
```

```python
print("Inertia : ")
print(round(kmeans.inertia_, 2))
```

```python
for k_test in [2,3,4,5]:
    km = KMeans(n_clusters=k_test, random_state = 42, n_init=10)
    km.fit(X)
    print(f" k = {k_test}, inertia = {round(kmeans.inertia_, 2)}")
```
