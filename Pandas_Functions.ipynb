```python
import pandas as pd
df= pd.read_csv("ecommerce_sales_analytics_5000.csv", parse_dates=['order_date'])
```

```python
df.info()
```

```python
df['Payment_Type'] = df['payment_method'].map({
    'Card': 'Digital',
    'Wallet': 'Digital',
    'COD': 'Cash'
}).fillna('Other')
```

```python
df['Payment_Type'].unique() 
```

```python
df[['Payment_Type','payment_method']].head()
```

```python
df.groupby("Payment_Type").size()
```

```python
def revenue_category(row):
    if row["revenue"] > 500:
        return "High"
    elif row["revenue"] >= 200:
        return "Medium"
    else:
        return "Low"

df["revenue_category"] = df.apply(revenue_category, axis=1)
```

```python
df["revenue_category"].value_counts()
```

```python
# Select the 10 samples first, then sort them
sample_df = df[["revenue", "revenue_category"]].sample(10)

# Define the custom order
order = {"High": 0, "Medium": 1, "Low": 2}

# Sort based on the mapping
sorted_sample = sample_df.sort_values(
    by="revenue_category", 
    key=lambda x: x.map(order)
)

print(sorted_sample)
```

```python
def add_profit(data):
    data["profit"] = data["revenue"] * 0.20
    return data

df = ( 
    df
    .pipe(add_profit)
)
```
