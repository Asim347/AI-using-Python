```python
# Task 1
salary = float(input("Enter salary: "))

if salary < 30000:
    tax = salary * 0.05
elif salary <= 70000:
    tax = salary * 0.15
else:
    tax = salary * 0.25

final_salary = salary - tax

print("Tax:", tax)
print("Final Salary after tax:", final_salary)
```

```python
# task 3
class User:
    def __init__(self, name):
        self.name = name

    def join_chat(self, chatroom):
        chatroom.add_user(self)

    def leave_chat(self, chatroom):
        chatroom.remove_user(self)

    def send_message(self, chatroom, text):
        message = Message(self.name, text)
        chatroom.add_message(message)


class Message:
    def __init__(self, sender, text):
        self.sender = sender
        self.text = text


class ChatRoom:
    def __init__(self):
        self.users = []
        self.messages = []

    def add_user(self, user):
        self.users.append(user)
        print(user.name, "joined the chat")

    def remove_user(self, user):
        self.users.remove(user)
        print(user.name, "left the chat")

    def add_message(self, message):
        self.messages.append(message)
        print(message.sender + ": " + message.text)

    def view_history(self):
        print("Chat History:")
        for message in self.messages:
            print(message.sender + ": " + message.text)


chatroom = ChatRoom()

user1 = User("Ali")
user2 = User("Sara")

user1.join_chat(chatroom)
user2.join_chat(chatroom)

user1.send_message(chatroom, "Hello!")
user2.send_message(chatroom, "Hi Ali!")

chatroom.view_history()

user2.leave_chat(chatroom)
```
